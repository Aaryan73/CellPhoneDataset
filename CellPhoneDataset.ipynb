```
import pandas 
import requests 
```

```
df1_link = "https://raw.githubusercontent.com/zalihat/cell_phone_dataset/master/cell_phones.csv"    # Cell Phones Dataframe
df2_link = "https://raw.githubusercontent.com/zalihat/cell_phone_dataset/master/population1.csv"    # Population Dataframe
df3_link = "https://raw.githubusercontent.com/open-numbers/ddf--gapminder--systema_globalis/master/ddf--entities--geo--country.csv"    #Geo country Codes
```

```
link1 = requests.get(df1_link)
df1_link_content = link1.content
df1 = open('dataframe1.csv', 'wb')
df1.write(df1_link_content)
df1.close()
```

```
link2 = requests.get(df2_link)
df2_link_content = link2.content
df2 = open('dataframe2.csv', 'wb')
df2.write(df2_link_content)
df2.close()
```

```
link3 = requests.get(df3_link)
df3_link_content = link3.content
df3 = open('dataframe3.csv', 'wb')
df3.write(df3_link_content)
df3.close()
```

```
df1 = pandas.read_csv("dataframe1.csv")
df2 = pandas.read_csv("dataframe2.csv")
df3 = pandas.read_csv("dataframe3.csv")
```

```
df3.columns = ['geo', 'g77_and_oecd_countries', 'income_3groups', 'income_groups',
       'is--country', 'iso3166_1_alpha2', 'iso3166_1_alpha3',
       'iso3166_1_numeric', 'iso3166_2', 'landlocked', 'latitude', 'longitude',
       'main_religion_2008', 'Country', 'un_sdg_ldc', 'un_sdg_region', 'un_state',
       'unicef_region', 'unicode_region_subtag', 'world_4region',
       'world_6region']
```

```
df3[['geo', 'Country']]
```

```
merged_df = merged_df.merge(df3[['geo', 'Country']], on='Country')
```

```
merged_df['cell_phones_per_person'] = merged_df.cell_phones_total / merged_df.population
```

```
print("The cell phone to population ratio of USA in 2017 is: ", merged_df[merged_df.year == 2017][merged_df.Country=="United States"].cell_phones_per_person)
```

```
for i in range(len(merged_df['geo'])):
  merged_df['geo'][i]=merged_df['geo'][i].upper()
```
